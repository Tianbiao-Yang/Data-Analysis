```python
# input pakges
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
```

## pandas 数据结构

```python
# Series：包含一个值序列，并且包含数据标签（索引index）
obj = pd.Series([1, 2, 3, 4])
# 索引为0-(N-1)
obj
```

```python
# 获得索引
obj.values
obj.index
```

```python
# 创建单独的索引序列
obj2 = pd.Series([-1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
# 在自身改变
obj.index = ['w', 'b', 'c', 'd']
print(obj)
print(obj2)
```

```python
# Series 的name属性
obj2.name = 'data'
print(obj2)
obj2.index.name = 'state'
print(obj2)
```

```python
sdata = {'aa': 100, 'ww':12, 'ee': 13, 'qq':10}
sdata
```

```python
obj3 = pd.Series(sdata)
obj3
```

```python
# 你可以根据索引文件找对应的值
states = ['ff', 'aa', 'qq', 'ee']
obj4 = pd.Series(sdata, index=states)
obj4
```

```python
# isnull和notnull来检查缺失值
pd.isnull(obj4)
```

```python
pd.notnull(obj4)
```
